```python
%matplotlib inline

import matplotlib
from pandas import *
Stats = read_csv("2014_2015_EMS_SQR_Results_2015_11_17.csv")
Stats['Code'] = Stats['DBN'].map(lambda x: x[:2])
Stats
```

```python
Stats['Is Charter'] = Stats['Code'].apply(lambda x: 1 if x == '84' else 0)
```

```python
Stats['Is Charter'].hist()
```

```python
Stats['Percent Students with Disabilities'] = Stats['Percent Students with Disabilities'].map(lambda x: float(x.rstrip('%')))
Stats['Percent English Language Learners'] = Stats['Percent English Language Learners'].map(lambda x: float(x.rstrip('%')) if isinstance(x, str) else 0.0)
```

```python
Stats.dropna()
```
